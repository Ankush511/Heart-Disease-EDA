```python
df = pd.read_csv('heart.csv')
df.head()
```

```python
df.shape
```

## Correlation - Heatmap

```python
cor = df.corr(numeric_only=True)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(cor, annot=True)
```

## Correlation - Jointplot

```python
sns.jointplot(data=df, x='Age', y='MaxHR')
```

```python
sns.jointplot(data=df, x='Age', y='MaxHR', kind='hex')
```

```python
sns.jointplot(data=df, x='Age', y='MaxHR', kind='reg')
```

```python
sns.jointplot(data=df, x='HeartDisease', y='MaxHR', kind='reg')
```

The analysis indicates a negative correlation between Age and MaxHR, suggesting a decrease in maximum heart rate with age. Additionally, a positive correlation exists between Oldpeak and Heart Disease, implying that higher Oldpeak values are associated with an increased likelihood of having heart disease.
